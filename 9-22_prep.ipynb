```python
endpoint = 'https://api.twitter.com/2/tweets/search/recent'
```

```python
header = {'Authorization':'Bearer {}'.format(twitter_keys['Bearer_Token'].iloc[0])}
```

```python
header
```

```python
query = urllib.parse.quote('monkeypox -is:retweet lang:en')
```

```python
query
```

```python
api_url = endpoint + '?query={}'.format(query)
```

```python
tweet_fields = 'id,created_at,author_id,lang,public_metrics,context_annotations'
```

```python
api_url = endpoint + '?query={}&tweet.fields={}'.format(query, tweet_fields)
```

```python
api_url
```

```python
expansions = 'author_id'
```

```python
api_url = endpoint + '?query={}&tweet.fields={}&expansions={}'.format(query, tweet_fields, expansions)
```

```python
user_fields = 'username'
```

```python
api_url = endpoint + '?query={}&tweet.fields={}&expansions={}&user.fields={}'.format(query, tweet_fields, expansions, user_fields)
```

```python
response = requests.get(api_url, headers= header)
```

```python
response
```

```python
response.text
```

```python
response_dict = json.loads(response.text)
```

```python
response_dict.keys()
```

```python
response_dict['data']
```

```python
tweet_df = pd.DataFrame(response_dict['data'])
```

```python
tweet_df
```

```python
response_dict['includes']
```

```python
user_df = pd.DataFrame(response_dict['includes']['users'])
```

```python
full_df = pd.merge(tweet_df, user_df, how = 'inner', left_on = 'author_id', right_on = 'id')
```
