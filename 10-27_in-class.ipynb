```python
q_artists_albums_tracks = '''
SELECT Title as Album,
       a.Name as Artist,
       t.Name as Track
FROM albums r
INNER JOIN artists a
ON (r.ArtistId = a.ArtistId)
INNER JOIN tracks t
ON (r.AlbumId = t.AlbumId)
WHERE a.Name == "AC/DC"
'''
```

```python
artists_albums_tracks = pd.read_sql_query(q_artists_albums_tracks, chinook)
```

```python
artists_albums_tracks
```

```python
q_artists = '''
SELECT Name, ArtistId
FROM artists
'''
q_albums = '''
SELECT Title, ArtistId, AlbumId
FROM albums
'''
q_tracks = '''
SELECT Name, AlbumId
FROM tracks
'''

artists_df = pd.read_sql_query(q_artists, chinook)
albums_df = pd.read_sql_query(q_albums, chinook)
tracks_df = pd.read_sql_query(q_tracks, chinook)
```

```python
combined_df = artists_df.merge(albums_df, 'inner', left_on='ArtistId', right_on = 'ArtistId')
```

```python
combined_df = artists_df.merge(albums_df, 'inner', on = 'ArtistId')
```

```python
combined_df
```

```python
combined_df = combined_df.merge(tracks_df, 'inner', on = 'AlbumId')
```
